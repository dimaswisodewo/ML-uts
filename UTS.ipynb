```python
import pandas as pd
import numpy as np

features = ['MDVP:Fo(Hz)','MDVP:Fhi(Hz)','MDVP:Flo(Hz)',
              'MDVP:Jitter(%)','MDVP:RAP','MDVP:PPQ','Jitter:DDP',
              'MDVP:Shimmer','Shimmer:DDA','NHR','HNR',
              'RPDE','DFA','spread1','spread2','D2','PPE']
rawParkinsonData = pd.read_csv('parkinsons.csv')
parkinsonData = pd.read_csv('parkinsons.csv', usecols=[features[0],
                                                      features[1],
                                                      features[2],
                                                      features[3],
                                                      features[4],
                                                      features[5],
                                                      features[6],
                                                      features[7],
                                                      features[8],
                                                      features[9],
                                                      features[10],
                                                      features[11],
                                                      features[12],
                                                      features[13],
                                                      features[14],
                                                      features[15],
                                                      features[16]])
```

```python
import matplotlib.pyplot as plt
import time
from pandas.plotting import scatter_matrix

scatter_matrix(parkinsonData, figsize=(20, 20))
plt.show()
```

```python
print(parkinsonData.corr(method='pearson'))
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as py
import time

start_time = time.time()

kmeans = KMeans(n_clusters=3, init='k-means++')
kmeans.fit(parkinsonData)

y_kmeans = kmeans.predict(parkinsonData)
centroid = kmeans.cluster_centers_

#X = np.array(parkinsonData[['Jitter:DDP','MDVP:Jitter(%)']])\
X = np.array(parkinsonData[['MDVP:RAP','MDVP:Jitter(%)']])

plt.figure(figsize=(10, 10), dpi=80)

plt.scatter(
    X[:, 0], X[:, 1],
    c=y_kmeans, s=150,
    edgecolor='blue', cmap='rainbow', alpha=.5)

#plt.scatter(
#    centroid[:, 0], centroid[:, 1],
#    c='yellow', s=50,
#    edgecolor='black', cmap='rainbow')

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
from sklearn import metrics

silhouetteScore = metrics.silhouette_score(X, y_kmeans, metric='euclidean', sample_size=300)

print("Silhouette Score: ", silhouetteScore)
```
